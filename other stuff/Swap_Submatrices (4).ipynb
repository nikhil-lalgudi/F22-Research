## Swap Submatrix program

Author: Tim Chartier

Description: Take two images (of the same size) and swap submatrices within the two images. 
 
Note: Images must be the same size in this version of the code. 


```python
# Set up parameters
cornerSize = 100;     # What's the size of a submatrix
numberOfPatches = 10; # number of submatrices will be swapped
```

```python
from PIL import Image
import numpy as np
from random import randint
```

```python
# Read the images 
img1  = Image.open('biden.png');
img2  = Image.open('trump.png');
```

```python
display(img1,img2)
```

```python
submatrixSize = 125
numberOfSwappedSubmatrices = 15
```

```python
# retrieve size of image
width, height = img1.size
```

```python
imageArray1 = np.asarray(img1)
imageArray2 = np.asarray(img2)
```

```python
imageSwapArray1 = np.copy(imageArray1)
imageSwapArray2 = np.copy(imageArray2)

for _ in range(numberOfSwappedSubmatrices):
    rowIndex = randint(0, height-submatrixSize-1)
    colIndex = randint(0, width-submatrixSize-1)
    imageSwapArray1[rowIndex:rowIndex+submatrixSize,colIndex:colIndex+submatrixSize] = imageArray2[rowIndex:rowIndex+submatrixSize,colIndex:colIndex+submatrixSize]
    imageSwapArray2[rowIndex:rowIndex+submatrixSize,colIndex:colIndex+submatrixSize] = imageArray1[rowIndex:rowIndex+submatrixSize,colIndex:colIndex+submatrixSize]
# Convert back to an image 
imgSwap1 = Image.fromarray(imageSwapArray1)
imgSwap2 = Image.fromarray(imageSwapArray2)
display(imgSwap1,imgSwap2)
```
